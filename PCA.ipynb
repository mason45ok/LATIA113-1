```python
import pandas as pd

# Path to your CSV file
file_path = 'A_stat_data_112.csv'

# Read the CSV file
data = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
data
```

```python
data.columns
```

```python
result = (
    data.groupby('user_id')  # 根據 'user_id' 分組
    .agg({
        'correct_rate': 'sum',                 # 將 'correct_rate' 加總
        'answer_duration_sec': 'sum',         # 將答題時間 'answer_duration_sec' 加總
        'user_id': 'size',                    # 計算該用戶的行數
        'mission_id': 'first',                # 保留第一筆任務 ID
        'grade': 'first',
        'is_completed':'sum'# 保留第一筆年級
    })
    .rename(columns={'user_id': 'count'})      # 重命名 'user_id' 的聚合欄位為 'count'
    .reset_index()
)

# 調整欄位順序
result = result[['user_id', 'correct_rate', 'answer_duration_sec', 'count', 'mission_id', 'grade']]
```

```python
result
```

```python
result['correct_rate'] = result['correct_rate'] / result['count']
result.head(1)
```

```python
# 為 answer_duration_sec_level 動態生成標籤
answer_bins = pd.qcut(
    result['answer_duration_sec'], 
    q=5, 
    duplicates='drop'
)
result['answer_duration_sec_level'] = pd.qcut(
    result['answer_duration_sec'], 
    q=5, 
    labels=range(1, len(answer_bins.cat.categories) + 1),  # 根據實際區間數生成標籤
    duplicates='drop'
)

# 為 count_level 動態生成標籤
count_bins = pd.qcut(
    result['count'], 
    q=5, 
    duplicates='drop'
)
result['count_level'] = pd.qcut(
    result['count'], 
    q=5, 
    labels=range(1, len(count_bins.cat.categories) + 1),  # 根據實際區間數生成標籤
    duplicates='drop'
)
```

```python
result.head(5)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import plotly.express as px

# 選擇需要的欄位
data = result[['answer_duration_sec_level', 'correct_rate', 'count_level', 'grade']]

# 移除缺失值
data = data.dropna()

# 標準化數據
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# 應用 PCA 將數據降到 3 維
pca = PCA(n_components=3)
pca_data = pca.fit_transform(scaled_data)
pca_df = pd.DataFrame(pca_data, columns=['PCA1', 'PCA2', 'PCA3'])

# K-means 聚類，分成三類
kmeans = KMeans(n_clusters=3, random_state=0)
pca_df['Cluster'] = kmeans.fit_predict(pca_data)

# 添加 ID 以便追蹤，並加入原數據以方便後續分析
pca_df[['answer_duration_sec_level', 'correct_rate', 'count_level', 'grade']] = data.reset_index(drop=True)

# 繪製 3D 結果
fig = px.scatter_3d(
    pca_df,
    x='PCA1',
    y='PCA2',
    z='PCA3',
    color='Cluster',
    hover_data={
        'PCA1': False,
        'PCA2': False,
        'PCA3': False,
        'Cluster': True,
        'answer_duration_sec_level': True,
        'count_level': True,
        'correct_rate': True,
        'grade': True
    },
    title="3D PCA and K-means Clustering of Game Time, Correct Ratio, Count, and Chinese Score"
)
fig.update_layout(
    scene=dict(
        xaxis_title="Principal Component 1",
        yaxis_title="Principal Component 2",
        zaxis_title="Principal Component 3"
    )
)
fig.show()
```
